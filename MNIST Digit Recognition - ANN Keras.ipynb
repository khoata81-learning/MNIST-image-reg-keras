# 1. Data Preparation  

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
```

## 1.1 Loading data 

```python
df = pd.read_csv('train.csv')
df
```

```python
df_train, df_test = train_test_split(df, test_size=0.2)
```

## 1.2 Split X & y 

### - Train dataset 

```python
ytrain = df_train['label']
Xtrain = df_train.drop(['label'], axis=1)
```

### - TEST dataset

```python
ytest = df_test['label']
Xtest = df_test.drop(['label'], axis=1)
```

```python
ytest
```

```python
Xtest
```

## 1.3 Normalize data [0..255] to [0..1] !!! IMPORTANT 

This step is very important, so that when we do validation or prediction with new data, the model can recognize patterns 

```python
Xtrain = Xtrain/255
Xtrain
```

```python
Xtest = Xtest/255
Xtest
```

## 1.4 Preview image 

```python
import matplotlib.pyplot as plt 
```

```python
def preview_image(flat_pixel, label_true=-1, label_pred=-1):
    ''' Input
        - flat_pixel: numpy 1D array, size = 784
    '''
    print('- Label_true:', label_true)
    print('- Label_predict:', label_pred)
    image = flat_pixel.reshape(28, 28)
    plt.imshow(image, cmap='gray')
    plt.show() 
```

```python
preview_image(Xtrain.iloc[0].to_numpy())
```

# 2. Model building using Keras 

## 2.1 Setup model architect

```python
from tensorflow import keras 
```

```python
# Layer 1 = 128 neurons = 2^7 so model can learn the pattern round a point + utilizing the power of CPU 2^n
# Activation ReLU: keep color values >= 0, don't use softmax as it encrypt data to 0 and 1, data can be lost and gradient is vanished 

nn_model = keras.Sequential([
    keras.layers.InputLayer(shape=(Xtrain.shape[-1],)), 
    keras.layers.Dense(units=128, activation=keras.activations.relu), 
    keras.layers.Dense(units=64, activation=keras.activations.relu), 
    keras.layers.Dense(units=32, activation=keras.activations.relu), 
    keras.layers.Dense(units=10, activation=keras.activations.softmax) # output layer: digit from 0..9
])
```

```python
nn_model.summary()
```

## 2.2 Compile the model 

```python
# Test with Adam

nn_model.compile(
    optimizer=keras.optimizers.SGD(), 
    loss=keras.losses.CategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
# Convert ytrain to categorical with 10 cols (~ 10 neurons), so that Keras can compare the output with ytrain_true 
ytrain = keras.utils.to_categorical(ytrain, num_classes=10)
ytrain
```

```python
Xtrain
```

```python
# Convert ytrain to categorical with 10 cols (~ 10 neurons), so that Keras can compare the output with ytrain_true 
ytest = keras.utils.to_categorical(ytest, num_classes=10)
ytest
```

## 2.3 Train the model 

```python
nn_model.fit(Xtrain, ytrain, epochs=10, validation_data=(Xtest, ytest)) 
```

# 3. Validation with completely new dataset 

```python
df_val = pd.read_csv('test.csv')
df_val
```

```python
# Normalize data 
Xval = df_val/255
Xval
```

```python
yval_pred = nn_model.predict(Xval).argmax(axis=1)
```

```python
for i in range(20):
    preview_image(Xval.iloc[i].to_numpy(), label_pred=yval_pred[i])
```
